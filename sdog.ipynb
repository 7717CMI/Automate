```python
import requests
import json

api_key = "686e26cc8d1ab67366cf2af2"
base_url = "https://api.scrapingdog.com/google_trends"

def get_trends_data(keyword):
    """
    Get Interest Over Time and Interest By Region data for a keyword
    Filters: Worldwide, 7 days analysis
    """
    results = {}
    
    # Common parameters for worldwide, 7 days
    base_params = {
        "api_key": api_key,
        "geo": "",  # Worldwide
        "query": keyword,
        "tz": "360",
        "date": "now 7-d",  # 7 days analysis
    }
    
    # Get Interest Over Time
    timeseries_params = {**base_params, "data_type": "TIMESERIES"}
    response = requests.get(base_url, params=timeseries_params)
    
    if response.status_code == 200:
        data = response.json()
        results['interest_over_time'] = data.get('interest_over_time', {})
    else:
        print(f"Failed to get timeseries data for '{keyword}': {response.status_code}")
        return None
    
    # Get Interest By Region
    geo_params = {**base_params, "data_type": "GEO_MAP"}
    response = requests.get(base_url, params=geo_params)
    
    if response.status_code == 200:
        data = response.json()
        results['interest_by_region'] = data.get('compared_breakdown_by_region', data.get('interest_by_region', {}))
    else:
        print(f"Failed to get regional data for '{keyword}': {response.status_code}")
        return None
    
    return results

def check_keyword_qualification(keyword, data):
    """
    Check if keyword meets qualification criteria:
    a) Interest over time should cross value 50 at least twice
    b) Interest by region should contain at least one target country
    """
    target_countries = ['United States', 'United Kingdom', 'China', 'India', 
                       'South Korea', 'France', 'Spain', 'South Africa', 'Canada']
    
    # Check Interest Over Time criteria
    interest_over_time = data.get('interest_over_time', {})
    timeline_data = interest_over_time.get('timeline_data', [])
    
    values_above_50 = 0
    for entry in timeline_data:
        values = entry.get('values', [])
        for value_data in values:
            if value_data.get('query', '').strip() == keyword.strip():
                try:
                    value = int(value_data.get('value', 0))
                    if value >= 50:
                        values_above_50 += 1
                except (ValueError, TypeError):
                    continue
    
    time_criteria_met = values_above_50 >= 2
    
    # Check Interest By Region criteria
    interest_by_region = data.get('interest_by_region', {})
    region_criteria_met = False
    
    # Handle different possible structures for regional data
    if isinstance(interest_by_region, list):
        regions = interest_by_region
    elif isinstance(interest_by_region, dict):
        regions = interest_by_region.get('default', {}).get('geoMapData', [])
        if not regions:
            regions = list(interest_by_region.values())
    else:
        regions = []
    
    # Check if any target country is present
    for region in regions:
        if isinstance(region, dict):
            region_name = region.get('geoName', region.get('location', ''))
            if any(country.lower() in region_name.lower() for country in target_countries):
                region_criteria_met = True
                break
    
    return {
        'keyword': keyword,
        'qualified': time_criteria_met and region_criteria_met,
        'time_criteria_met': time_criteria_met,
        'values_above_50_count': values_above_50,
        'region_criteria_met': region_criteria_met,
        'time_criteria_details': f"Found {values_above_50} values >= 50 (need at least 2)",
        'region_criteria_details': f"Target country found: {region_criteria_met}"
    }

def analyze_keywords(keywords_list):
    """
    Main function to analyze multiple keywords and return qualified ones
    """
    qualified_keywords = []
    all_results = []
    
    print(" Analyzing Keywords...")
    print("=" * 60)
    
    for keyword in keywords_list:
        print(f"\n Processing: '{keyword}'")
        print("-" * 40)
        
        # Get trends data
        data = get_trends_data(keyword)
        
        if data is None:
            print(f" Failed to get data for '{keyword}'")
            continue
        
        # Check qualification
        result = check_keyword_qualification(keyword, data)
        all_results.append(result)
        
        # Display results
        print(f" Time Criteria: {result['time_criteria_details']}")
        print(f" Region Criteria: {result['region_criteria_details']}")
        print(f" Qualified: {'YES' if result['qualified'] else 'NO'}")
        
        if result['qualified']:
            qualified_keywords.append(keyword)
    
    # Summary
    print("\n" + "=" * 60)
    print(" FINAL RESULTS")
    print("=" * 60)
    
    print(f"\n QUALIFIED KEYWORDS ({len(qualified_keywords)}):")
    if qualified_keywords:
        for i, keyword in enumerate(qualified_keywords, 1):
            print(f"  {i}. {keyword}")
    else:
        print("  No keywords met all criteria")
    
    print(f"\n DETAILED BREAKDOWN:")
    for result in all_results:
        status = " QUALIFIED" if result['qualified'] else " NOT QUALIFIED"
        print(f"   {result['keyword']}: {status}")
        print(f"    - Time: {result['time_criteria_details']}")
        print(f"    - Region: {result['region_criteria_details']}")
    
    return qualified_keywords, all_results

# Example usage
if __name__ == "__main__":
    # Test with your keywords
    test_keywords = [
        "AI technology",
        "Network",
        "Python programming",
        "Machine learning"
    ]
    
    print(" ScrapingDog Google Trends Keyword Analyzer")
    print(" Filter: Worldwide, 7 days")
    print(" Criteria: Interest>50 (2x) + Target Countries")
    print("=" * 60)
    
    qualified, detailed_results = analyze_keywords(test_keywords)
    
    print(f"\n Analysis Complete!")
    print(f"   Processed: {len(test_keywords)} keywords")
    print(f"   Qualified: {len(qualified)} keywords")
```

```python
import requests
import json

api_key = "686e263c4bbdf9e398881690"
base_url = "https://api.scrapingdog.com/google_trends"

def get_trends_data(keyword):
    """
    Get Interest Over Time and Interest By Region data for a keyword
    Filters: Worldwide, 7 days analysis
    """
    results = {}
    
    # Clean and format the keyword
    clean_keyword = keyword.strip()
    
    # Try different date formats if 7-day doesn't work
    date_options = ["now 7-d", "today 1-m", "now 1-d"]
    
    for date_option in date_options:
        # Common parameters for worldwide
        base_params = {
            "api_key": api_key,
            "geo": "",  # Worldwide
            "query": clean_keyword,
            "tz": "360",
            "date": date_option,
        }
        
        # Get Interest Over Time
        timeseries_params = {**base_params, "data_type": "TIMESERIES"}
        response = requests.get(base_url, params=timeseries_params)
        
        if response.status_code == 200:
            data = response.json()
            results['interest_over_time'] = data.get('interest_over_time', {})
            print(f"   Got timeseries data with date: {date_option}")
            break
        else:
            print(f"   Timeseries failed with {date_option}: {response.status_code}")
            if response.status_code == 400:
                try:
                    error_data = response.json()
                    print(f"     Error details: {error_data}")
                except:
                    print(f"     Error response: {response.text[:200]}")
    else:
        print(f"Failed to get timeseries data for '{keyword}' with all date options")
        return None
    
    # Get Interest By Region using the same successful date option
    successful_date = date_option
    geo_params = {
        "api_key": api_key,
        "geo": "",
        "query": clean_keyword,
        "tz": "360",
        "date": successful_date,
        "data_type": "GEO_MAP"
    }
    
    response = requests.get(base_url, params=geo_params)
    
    if response.status_code == 200:
        data = response.json()
        results['interest_by_region'] = data.get('compared_breakdown_by_region', data.get('interest_by_region', {}))
        print(f"   Got regional data")
    else:
        print(f"   Regional data failed: {response.status_code}")
        if response.status_code == 400:
            try:
                error_data = response.json()
                print(f"     Error details: {error_data}")
            except:
                print(f"     Error response: {response.text[:200]}")
        # Continue anyway with just timeseries data
        results['interest_by_region'] = {}
    
    return results

def check_keyword_qualification(keyword, data):
    """
    Check if keyword meets qualification criteria:
    a) Interest over time should cross value 50 at least twice
    b) Interest by region should contain at least one target country
    """
    target_countries = ['United States', 'United Kingdom', 'China', 'India', 
                       'South Korea', 'France', 'Spain', 'South Africa', 'Canada']
    
    # Check Interest Over Time criteria
    interest_over_time = data.get('interest_over_time', {})
    timeline_data = interest_over_time.get('timeline_data', [])
    
    values_above_50 = 0
    for entry in timeline_data:
        values = entry.get('values', [])
        for value_data in values:
            if value_data.get('query', '').strip() == keyword.strip():
                try:
                    value = int(value_data.get('value', 0))
                    if value >= 50:
                        values_above_50 += 1
                except (ValueError, TypeError):
                    continue
    
    time_criteria_met = values_above_50 >= 2
    
    # Check Interest By Region criteria
    interest_by_region = data.get('interest_by_region', {})
    region_criteria_met = False
    
    # Handle different possible structures for regional data
    if isinstance(interest_by_region, list):
        regions = interest_by_region
    elif isinstance(interest_by_region, dict):
        regions = interest_by_region.get('default', {}).get('geoMapData', [])
        if not regions:
            regions = list(interest_by_region.values())
    else:
        regions = []
    
    # Check if any target country is present
    for region in regions:
        if isinstance(region, dict):
            region_name = region.get('geoName', region.get('location', ''))
            if any(country.lower() in region_name.lower() for country in target_countries):
                region_criteria_met = True
                break
    
    return {
        'keyword': keyword,
        'qualified': time_criteria_met and region_criteria_met,
        'time_criteria_met': time_criteria_met,
        'values_above_50_count': values_above_50,
        'region_criteria_met': region_criteria_met,
        'time_criteria_details': f"Found {values_above_50} values >= 50 (need at least 2)",
        'region_criteria_details': f"Target country found: {region_criteria_met}"
    }

def analyze_keywords(keywords_list):
    """
    Main function to analyze multiple keywords and return qualified ones
    """
    qualified_keywords = []
    all_results = []
    
    print(" Analyzing Keywords...")
    print("=" * 60)
    
    for keyword in keywords_list:
        print(f"\n Processing: '{keyword}'")
        print("-" * 40)
        
        # Get trends data
        data = get_trends_data(keyword)
        
        if data is None:
            print(f" Failed to get data for '{keyword}'")
            continue
        
        # Check qualification
        result = check_keyword_qualification(keyword, data)
        all_results.append(result)
        
        # Display results
        print(f" Time Criteria: {result['time_criteria_details']}")
        print(f" Region Criteria: {result['region_criteria_details']}")
        print(f" Qualified: {'YES' if result['qualified'] else 'NO'}")
        
        if result['qualified']:
            qualified_keywords.append(keyword)
    
    # Summary
    print("\n" + "=" * 60)
    print(" FINAL RESULTS")
    print("=" * 60)
    
    print(f"\n QUALIFIED KEYWORDS ({len(qualified_keywords)}):")
    if qualified_keywords:
        for i, keyword in enumerate(qualified_keywords, 1):
            print(f"  {i}. {keyword}")
    else:
        print("  No keywords met all criteria")
    
    print(f"\n DETAILED BREAKDOWN:")
    for result in all_results:
        status = " QUALIFIED" if result['qualified'] else " NOT QUALIFIED"
        print(f"   {result['keyword']}: {status}")
        print(f"    - Time: {result['time_criteria_details']}")
        print(f"    - Region: {result['region_criteria_details']}")
    
    return qualified_keywords, all_results

# Example usage
if __name__ == "__main__":
    # Test with your keywords
    test_keywords = [
        "Big Data Healthcare",
        "Interventional Neurology",
        "Prosthetic Heart Valve",
        "Machine learning"
    ]
    
    print(" ScrapingDog Google Trends Keyword Analyzer")
    print(" Filter: Worldwide, 7 days")
    print(" Criteria: Interest>50 (2x) + Target Countries")
    print("=" * 60)
    
    qualified, detailed_results = analyze_keywords(test_keywords)
    
    print(f"\n Analysis Complete!")
    print(f"   Processed: {len(test_keywords)} keywords")
    print(f"   Qualified: {len(qualified)} keywords")
```
